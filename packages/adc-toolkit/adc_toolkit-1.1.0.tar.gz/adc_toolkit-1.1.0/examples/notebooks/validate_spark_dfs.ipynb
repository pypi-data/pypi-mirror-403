```python
import os


# Set JAVA_HOME to Java 17 for PySpark compatibility (Java 24+ is not supported)
os.environ["JAVA_HOME"] = "/opt/homebrew/opt/openjdk@17/libexec/openjdk.jdk/Contents/Home"

from pyspark.sql import SparkSession

from adc_toolkit.data.catalog import ValidatedDataCatalog
from adc_toolkit.data.validators.gx import GXValidator
from adc_toolkit.data.validators.no_validator import NoValidator
from adc_toolkit.data.validators.pandera import PanderaValidator
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
path = os.path.abspath("./examples/configs/catalog")
```

```python
catalog = ValidatedDataCatalog.in_directory(path, validator_class=NoValidator)
```

```python
df = catalog.load("elnino.extract")
```

```python
df_spark = spark.createDataFrame(df)
```

```python
gx_validator = GXValidator.in_directory(path)
```

```python
df_spark_validated_gx = gx_validator.validate("elnino.extract", df_spark)
```

```python
pandera_validator = PanderaValidator.in_directory(path)
```

```python
df_spark_validated_pandera = pandera_validator.validate("elnino.extract", df_spark)
```
