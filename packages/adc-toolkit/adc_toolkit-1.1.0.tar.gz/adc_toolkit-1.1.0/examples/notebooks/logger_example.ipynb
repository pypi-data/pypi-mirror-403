### Using Logger
If you don't have loguru, it will use the Python logger implementation.

```python
from importlib.util import find_spec

from adc_toolkit.logger import Logger
from examples.scripts.logging_example.main import main
```

```python
logger = Logger() if find_spec("loguru") else Logger(__name__)

logger.debug("debugging...")
logger.info("information")
logger.warning("warning")
logger.error("error")
```

#### Setting a log level within your module

```python
Logger.set_level("info")
logger = Logger(__name__)

logger.debug("This message will not show")

Logger.set_level("debug")
logger = Logger(__name__)

logger.debug("This message will show")
```

#### Setting custom options

```python
# Set an option
Logger.set_options(format_name="CUSTOM NAME")
logger = Logger(__name__)

logger.warning("This message will show a custom name")

# Reset the options to the defaults
Logger.reset_options()
logger = Logger(__name__)

logger.warning("This message will not show a custom name")
```
