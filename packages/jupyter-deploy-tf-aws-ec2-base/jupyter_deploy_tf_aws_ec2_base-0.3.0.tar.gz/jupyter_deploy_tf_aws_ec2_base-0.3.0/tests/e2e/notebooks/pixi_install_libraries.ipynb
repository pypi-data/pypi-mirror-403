# Pixi Package Manager Test - Install Libraries

Test installing libraries via Pixi package manager (both PyPI and conda).

```python
# Install pytest from PyPI via pixi
!pixi add --pypi pytest
```

```python
import subprocess

import pytest
```

```python
# Verify pytest is functional
version = pytest.__version__
print(f"pytest version: {version}")
assert version, "pytest version should not be empty"
print("✓ pytest installed and functional")
```

```python
# Install conda-build from conda (not available on PyPI)
!pixi add conda-build
```

```python
# Verify conda-build is available via pixi list
result = subprocess.run(["pixi", "list"], capture_output=True, text=True, check=True)

# Check that conda-build is in the list
assert "conda-build" in result.stdout, "conda-build not found in pixi list"
print("✓ conda-build is installed")
```

```python
# Verify conda-build command is functional
result = subprocess.run(["conda-build", "--version"], capture_output=True, text=True, check=True)
version = result.stdout.strip()
print(f"conda-build version: {version}")
assert version, "conda-build version should not be empty"
print("✓ conda-build is functional")
```
