# ModSSC | HPO Bench Smoke

Run the benchmark with a hyperparameter search config and inspect the best trial.

## Objective
- Execute the bench runner on the toy HPO config.
- Read the HPO summary and best trial from disk.


```python
import subprocess
from pathlib import Path

config_path = Path("bench/configs/experiments/toy_inductive_hpo.yaml")
result = subprocess.run(
    ["python", "-m", "bench.main", "--config", str(config_path)],
    capture_output=True,
    text=True,
)
print(result.stdout)
print(result.stderr)
result.returncode
```

```python
import json

runs_dir = Path("runs")
run_dirs = sorted(runs_dir.glob("toy_inductive_hpo-*"))
if not run_dirs:
    raise FileNotFoundError("No runs found for toy_inductive_hpo")
run_dir = run_dirs[-1]
run = json.loads((run_dir / "run.json").read_text())
hpo = run.get("hpo", {})
hpo
```

```python
trials_path = run_dir / hpo["trials_path"]
trials = [json.loads(line) for line in trials_path.read_text().splitlines()]
best = next(t for t in trials if t["index"] == hpo["best_index"])
best
```
