```python
import requests
from io import StringIO
from src.request.get import get_url
import pandas as pd

def get_proxies() -> list:
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
    }
    url= "https://spys.one/en/free-proxy-list/"
    response_body = requests.get(url, headers=headers)
    tables = pd.read_html(StringIO(response_body.text))
    ip_table = tables
    return ip_table
    proxies_list = ip_table.apply(
    lambda row: {row['Type'].lower(): f"{row['IP Address']}:{row['Port']}"},
    axis=1
    ).tolist()
    return proxies_list

if __name__=='__main__':
    proxy_list = get_proxies()
    print(proxy_list)
```

Testing SQLite with SQL Alchemy

```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import declarative_base, sessionmaker
```

```python
engine = create_engine("sqlite:///example.db" , echo=True)
Base = declarative_base()
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

```python
class User(Base):
    __tablename__ = "users"

    id = Column(Integer, primary_key=True)
    name = Column(String, nullable=False)
    email = Column(String, unique=True)

Base.metadata.create_all(engine)
```

```python
user = User(name="Praful", email="praful@email.com")
session.add(user)
session.commit()
```

```python
users = session.query(User).all()
for user in users:
    print(user.id, user.name, user.email)
```
