```python
from os import getcwd
from os.path import join, dirname
from FLiESANN import load_ECOv002_calval_FLiESANN_inputs
from BESS_JPL import load_ECOv002_calval_BESS_inputs
from STIC_JPL import load_ECOv002_calval_STIC_inputs
from PMJPL import load_ECOv002_calval_PMJPL_inputs
from PTJPLSM import load_ECOv002_calval_PTJPLSM_inputs
from ECOv003_L3T_L4T_JET import process_JET_table
```

```python
repo_root = dirname(getcwd())
package_dir = join(repo_root, 'ECOv003_L3T_L4T_JET')
generated_input_table_filename = join(package_dir, "ECOv002-cal-val-JET-inputs.csv")
generated_output_table_filename = join(package_dir, "ECOv002-cal-val-JET-outputs.csv")
```

```python
FLiES_df = load_ECOv002_calval_FLiESANN_inputs()
FLiES_df.head()
```

```python
inputs_df = FLiES_df.copy()
inputs_df.head()
```

```python
BESS_df = load_ECOv002_calval_BESS_inputs()
BESS_df.head()
```

```python
sorted(BESS_df.columns)
```

```python
for col in BESS_df.columns:
    if col not in inputs_df.columns:
        inputs_df[col] = BESS_df[col]

inputs_df.head()
```

```python
STIC_df = load_ECOv002_calval_STIC_inputs()
STIC_df.head()
```

```python
for col in STIC_df.columns:
    if col not in inputs_df.columns:
        inputs_df[col] = STIC_df[col]

inputs_df.head()
```

```python
PMJPL_df = load_ECOv002_calval_PMJPL_inputs()
PMJPL_df.head()
```

```python
for col in PMJPL_df.columns:
    if col not in inputs_df.columns:
        inputs_df[col] = PMJPL_df[col]

inputs_df.head()
```

```python
PTJPLSM_df = load_ECOv002_calval_PTJPLSM_inputs()
PTJPLSM_df.head()
```

```python
for col in PTJPLSM_df.columns:
    if col not in inputs_df.columns:
        inputs_df[col] = PTJPLSM_df[col]

inputs_df.head()
```

```python
# Convert time_UTC to ISO8601 format
import pandas as pd
inputs_df['time_UTC'] = pd.to_datetime(inputs_df['time_UTC']).dt.strftime('%Y-%m-%dT%H:%M:%SZ')
```

```python
inputs_df.to_csv(generated_input_table_filename, index=False)
```

```python
outputs_df = process_JET_table(inputs_df, offline_mode=True, include_water_surface=False)
outputs_df
```
