# Evaluation Integration Test - Clean Version

This notebook tests the evaluation functionality with proper package imports.

```python
import sys
from pathlib import Path

import bedrock_agentcore_starter_toolkit
from bedrock_agentcore_starter_toolkit import Evaluation

# Setup: Add local package to path
# Get repository root (goes up from notebook -> tests_integ -> repo root)
repo_root = Path.cwd().parent.parent
sys.path.insert(0, str(repo_root / "src"))

print(f"Added to path: {repo_root / 'src'}")

# Now import from local package
# Verify correct package
print(f"\n✅ Using package from: {bedrock_agentcore_starter_toolkit.__file__}")

# Verify get_latest_session exists
eval_test = Evaluation(region="us-east-1")
has_method = hasattr(eval_test, "get_latest_session")
print(f"✅ Has get_latest_session method: {has_method}")
if not has_method:
    print("\n❌ ERROR: Wrong package! Not using bedrock-agentcore-starter-toolkit-evaluation")
```

```python
# Configuration
TEST_AGENT_ID = "agent_lg-EVQuBO6Q0n"  # Replace with your agent ID
TEST_SESSION_ID = None  # Set to None to auto-fetch latest, or provide explicit session ID
TEST_REGION = "us-east-1"

print("Configuration:")
print(f"  Agent ID:    {TEST_AGENT_ID}")
print(f"  Session ID:  {TEST_SESSION_ID or '(will auto-fetch latest)'}")
print(f"  Region:      {TEST_REGION}")

# Note: Auto-fetch queries the last 7 days of sessions
# If your session is older, provide an explicit session ID
```

## Test 1: Initialize Evaluation Client

```python
eval_client = Evaluation(region=TEST_REGION)

print("✅ Test 1 PASSED: Evaluation client initialized")
print(f"  Region:   {eval_client.region}")
print("\nNote: agent_id is now passed as parameter to run() method, not stored in client")
```

## Test 2: List Evaluators

```python
response = eval_client.list_evaluators(max_results=50)
evaluators = response.get("evaluators", [])

builtin = [e for e in evaluators if e.get("evaluatorId", "").startswith("Builtin.")]
custom = [e for e in evaluators if not e.get("evaluatorId", "").startswith("Builtin.")]

print(f"✅ Test 2 PASSED: Found {len(evaluators)} evaluators")
print(f"  Builtin: {len(builtin)}")
print(f"  Custom:  {len(custom)}")

if builtin:
    print("\nSample builtin evaluators:")
    for ev in builtin[:3]:
        print(f"  - {ev.get('evaluatorId')}")
```

## Test 3: Get Evaluator Details

```python
details = eval_client.get_evaluator("Builtin.Helpfulness")

print("✅ Test 3 PASSED: Retrieved evaluator details")
print(f"  Name:  {details.get('evaluatorName')}")
print(f"  Level: {details.get('level')}")
print(f"  Description: {details.get('description')[:100]}...")
```

## Test 4: Run Evaluation (Auto-fetch Session)

This test will auto-fetch the latest session if TEST_SESSION_ID is None.

```python
results = eval_client.run(agent_id=TEST_AGENT_ID, session_id=TEST_SESSION_ID, evaluators=["Builtin.GoalSuccessRate"])

print("✅ Test 4 PASSED: Evaluation completed")
print(f"  Session ID: {results.session_id}")
print(f"  Results:    {len(results.results)}")

successful = results.get_successful_results()
failed = results.get_failed_results()

print(f"  Successful: {len(successful)}")
print(f"  Failed:     {len(failed)}")

if successful:
    result = successful[0]
    print("\n📊 Result:")
    print(f"  Evaluator: {result.evaluator_name}")
    print(f"  Score:     {result.value:.2f}")
    print(f"  Label:     {result.label}")
    if result.explanation:
        print(f"  Explanation: {result.explanation[:150]}...")
```

## Test 5: Run with Multiple Evaluators

```python
# Use the session from previous test
session_to_use = TEST_SESSION_ID or results.session_id

results = eval_client.run(
    agent_id=TEST_AGENT_ID, session_id=session_to_use, evaluators=["Builtin.Helpfulness", "Builtin.Faithfulness"]
)

print("✅ Test 5 PASSED: Multi-evaluator completed")
print(f"  Results: {len(results.results)}")

for result in results.get_successful_results():
    print(f"  - {result.evaluator_name}: {result.value:.2f} ({result.label})")
```

## Test 6: Export Results to JSON

```python
import tempfile

output_file = Path(tempfile.gettempdir()) / "eval_test_results.json"

results = eval_client.run(
    agent_id=TEST_AGENT_ID, session_id=session_to_use, evaluators=["Builtin.GoalSuccessRate"], output=str(output_file)
)

print("✅ Test 6 PASSED: Results exported")
print(f"  File: {output_file}")
print(f"  Size: {output_file.stat().st_size} bytes")

# Check for input data file
input_file = output_file.parent / f"{output_file.stem}_input{output_file.suffix}"
if input_file.exists():
    print(f"  Input file: {input_file} ({input_file.stat().st_size} bytes)")
```

## Test 7: Create Custom Evaluator

```python
import time

custom_config = {
    "llmAsAJudge": {
        "modelConfig": {
            "bedrockEvaluatorModelConfig": {
                "modelId": "global.anthropic.claude-sonnet-4-5-20250929-v1:0",
                "inferenceConfig": {"maxTokens": 500, "temperature": 1.0},
            }
        },
        "ratingScale": {
            "numerical": [
                {"value": 0.0, "label": "Verbose", "definition": "Response is overly wordy"},
                {"value": 1.0, "label": "Concise", "definition": "Response is concise"},
            ]
        },
        "instructions": "Evaluate response conciseness. Context: {context}. Target: {assistant_turn}",
    }
}

custom_name = f"test_eval_{int(time.time())}"

response = eval_client.create_evaluator(
    name=custom_name, config=custom_config, level="TRACE", description="Test evaluator for notebook"
)

CUSTOM_EVALUATOR_ID = response.get("evaluatorId")

print("✅ Test 7 PASSED: Custom evaluator created")
print(f"  ID:   {CUSTOM_EVALUATOR_ID}")
print(f"  Name: {custom_name}")
```

## Test 8: Run with Custom Evaluator

```python
if CUSTOM_EVALUATOR_ID:
    results = eval_client.run(agent_id=TEST_AGENT_ID, session_id=session_to_use, evaluators=[CUSTOM_EVALUATOR_ID])

    print("✅ Test 8 PASSED: Custom evaluator executed")
    print(f"  Results: {len(results.results)}")

    for result in results.get_successful_results():
        print(f"  Score: {result.value:.2f}, Label: {result.label}")

    for result in results.get_failed_results():
        print(f"  ❌ Error: {result.error[:100]}")
else:
    print("⚠️  Test 8 SKIPPED: No custom evaluator")
```

## Test 9: Update Custom Evaluator

```python
if CUSTOM_EVALUATOR_ID:
    response = eval_client.update_evaluator(
        CUSTOM_EVALUATOR_ID, description=f"Updated test evaluator - {int(time.time())}"
    )

    print("✅ Test 9 PASSED: Evaluator updated")
    if "updatedAt" in response:
        print(f"  Updated at: {response['updatedAt']}")
else:
    print("⚠️  Test 9 SKIPPED: No custom evaluator")
```

## Test 10: Delete Custom Evaluator (Cleanup)

```python
if CUSTOM_EVALUATOR_ID:
    eval_client.delete_evaluator(CUSTOM_EVALUATOR_ID)
    print("✅ Test 10 PASSED: Evaluator deleted")
    print(f"  Deleted: {CUSTOM_EVALUATOR_ID}")
else:
    print("⚠️  Test 10 SKIPPED: No custom evaluator")
```
