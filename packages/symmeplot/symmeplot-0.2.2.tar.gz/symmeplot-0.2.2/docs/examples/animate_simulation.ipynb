```python
import sympy as sm
import sympy.physics.mechanics as me
```

```python
q1, q2, u1, u2 = me.dynamicsymbols('q1 q2 u1 u2')
l, g, mc, mp = sm.symbols('l g mc mp')
N = me.ReferenceFrame('N')
O = me.Point('O')
O.set_vel(N, 0)

A = me.ReferenceFrame('A')
A.orient_axis(N, q1, N.z)
A.set_ang_vel(N, u1 * N.z)

cart = me.Particle('cart', me.Point("cart_mc"), mc)
pend = me.Particle('pend', me.Point("pend_mc"), mp)

cart.masscenter.set_pos(O, q2 * N.x)
pend.masscenter.set_pos(cart.masscenter, l * A.x)
```
